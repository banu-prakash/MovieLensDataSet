## importing all modules

```python
import pandas as pd

import plotly
from plotly.graph_objs import *
import plotly.offline as plot
import plotly.graph_objs as go
plot.offline.init_notebook_mode(connected=True)

import datetime
```

```python
u_user_df = pd.read_csv('u.user', 
                 names = ['user id', 'age', 'gender', 'occupation', 'zip code'], sep = '|', header = None)
u_data_df = pd.read_csv('u.data', 
                  names = ['user id', 'movie id', 'rating', 'timestamp'], sep = '\t', header = None)
u_item_df = pd.read_csv('u.item', 
                 names = ['movie id', 'movie title', 'release date', 'video release date', 'IMDb URL', 'unknown',
                          'Action', 'Adventure', 'Animation', 'Childrens', 'Comedy', 'Crime', 'Documentary', 
                          'Drama', 'Fantasy', 'Film Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci Fi',
                          'Thriller', 'War', 'Western'], sep = '|', header = None, encoding = "ISO-8859-1")
```

```python
Age = u_user_df.age
Rating = u_data_df.rating
```

## Distribution of Age Group

```python
data = [go.Histogram(x = Age)]
layout = dict(xaxis = dict(title = 'Age Range'),
              yaxis = dict(title = 'Frequency'),
              )
fig = dict(data = data, layout = layout)
plot.iplot(fig, filename='Age histogram')
```

## Rating Histogram

```python
data = [go.Histogram(x = Rating)]
layout = dict(xaxis = dict(title = 'Rating Range'),
              yaxis = dict(title = 'Frequency'),
              )
fig = dict(data = data, layout = layout)
plot.iplot(fig, filename='Rating histogram')
```

## Age Rating over the year

```python
u_data_df['timestamp'] = pd.to_datetime(u_data_df['timestamp'],unit='s')
u_data_df['Date'] = pd.DatetimeIndex(u_data_df['timestamp']).date
u_data_df.head()
```

```python
Average_rating_Date = pd.DataFrame(u_data_df.groupby(['Date']).rating.mean(), index=None)
```

```python
trace0 = Scatter(
        x = Average_rating_Date.index,
        y = Average_rating_Date.rating,
        #mode = 'markers'
)
data = [trace0]
layout = dict(xaxis = dict(title = 'Date'),
              yaxis = dict(title = 'Average Rating'),
              )
fig = dict(data = data, layout = layout)
plot.iplot(fig, filename='Rating histogram')
```

## Stacked BarPlot of Average Rating vs Age Group

```python
merge_data = pd.merge(u_data_df, u_user_df)
#merge_data.head()
```

```python
Average_rating_age = pd.DataFrame(merge_data.groupby(['age','gender']).rating.mean(), index=None)
Average_rating_age = Average_rating_age.unstack().dropna()
```

```python
Female_Rating = pd.DataFrame(Average_rating_age.rating['F'])
Male_Rating = pd.DataFrame(Average_rating_age.rating['M'])
xAxis = pd.DataFrame(Average_rating_age.index)
```

```python
trace0 = Bar(
        x = xAxis.age,
        y = Male_Rating['M'],
        name = 'Male'
)
trace1 = Bar(
        x = xAxis.age,
        y = Female_Rating['F'],
        name = 'Female'
)
data = [trace0, trace1]
layout = dict(
          xaxis = dict(title = 'Age'),
          yaxis = dict(title = 'Average Rating'),
          barmode='stack'
              )
fig = dict(data = data, layout = layout)
plot.iplot(fig, filename='Average rating vs Age Group')
```

## Occupation by Average Rating

```python
occupation_rating = pd.DataFrame(merge_data.groupby(['occupation']).rating.mean(), index=None)
```

```python
trace0 = Bar(
        x = occupation_rating.index,
        y = occupation_rating.rating,
        
)
data = [trace0]
layout = dict(xaxis = dict(title = 'Age'),
              yaxis = dict(title = 'Average Rating'),
              barmode='group'
              )
fig = dict(data = data, layout = layout)
plot.iplot(fig, filename='Average rating vs occupation')
```
